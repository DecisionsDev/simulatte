```python
#datasink_path = "/Users/tiemokodembele/Documents/internShip/simulatte/data/ADS/loanvalidation/run_my runner_20220615_182353.json"

outputData = []
if os.path.exists(datasink_path):
    with open(datasink_path) as file:
        filesize = os.path.getsize(datasink_path)
        if filesize == 0:
            print("The file {} is empty".format(datasink_path))
        else:
            while (line := file.readline().rstrip()):
                jsonLine = dict()
                line_to_json = json.loads(line)
                jsonLine["request"] = json.loads(line_to_json["request"])
                if(line_to_json["trace"]=="empty"): 
                    jsonLine["trace"] = "empty" 
                else: 
                    jsonLine["trace"] = json.loads(line_to_json["trace"])
                jsonLine["response"] = json.loads(line_to_json["response"])

                outputData.append(jsonLine)

datasink = pd.json_normalize(outputData)
#datasink = pd.json_normalize(outputData, ["taskInformation", "ruleInformation"], ["totalTasksNotExecuted", "shortname", ["tasksNotExecuted", "taskInformation"]])
datasink.head(15)
datasink.shape[0]
```

__BIDIRECTIONAL BAR CHARTS__

```python
import plotly.graph_objects as go

y=['number of applications', 'sum of applications amounts']

fig = go.Figure()
fig.add_trace(go.Bar(
    y=y,
    x=[-rejectedLoan['response.loan.amount'].shape[0], -rejectedLoan['response.loan.amount'].sum()],
    name='Rejected',
    base=0,
    orientation='h',
    marker=dict(
        color='rgba(246, 78, 139, 0.6)',
        line=dict(color='rgba(246, 78, 139, 1.0)', width=3)
    )
))
fig.add_trace(go.Bar(
    y=y,
    x=[approvedLoan['response.loan.amount'].shape[0], approvedLoan['response.loan.amount'].sum()],
    name='Accepted',
    base=0,
    orientation='h',
    marker=dict(
        color='rgba(58, 71, 80, 0.6)',
        line=dict(color='rgba(58, 71, 80, 1.0)', width=3)
    )
))

fig.update_layout(
    barmode='stack', 
    title={'text': f"Accetep vs Rejected",
          'x':0.5,
          'xanchor': 'center'
    })
fig.update_yaxes(
        ticktext=y,
        tickvals=y
    )
fig.show(auto_open=True)
plotly.offline.init_notebook_mode()
```

__BIDIRECTIONAL BAR CHARTS__ in 2 figs

```python
import plotly.graph_objects as go

y1=['sum of applications amounts']
y2=['number of applications']

fig1 = go.Figure()
fig2 = go.Figure()

fig1.add_trace(go.Bar(
    y=y1,
    x=[-rejectedLoan['response.loan.amount'].sum()],
    name='Rejected',
    base=0,
    #text=str(rejectedLoan['response.loan.amount'].sum()), 
    #hoverinfo ='text',
    orientation='h',
    marker=dict(
        color='rgba(246, 78, 139, 0.6)',
        line=dict(color='rgba(246, 78, 139, 1.0)', width=3)
    )
))

fig2.add_trace(go.Bar(
    y=y2,
    x=[-rejectedLoan['response.loan.amount'].shape[0]],
    name='Rejected',
    base=0,
    orientation='h',
    marker=dict(
        color='rgba(246, 78, 139, 0.6)',
        line=dict(color='rgba(246, 78, 139, 1.0)', width=3)
    )
))


fig1.add_trace(go.Bar(
    y=y1,
    x=[approvedLoan['response.loan.amount'].sum()],
    name='Accepted',
    base=0,
    #text= str(approvedLoan['response.loan.amount'].sum()), 
    #hoverinfo ='text',
    orientation='h',
    marker=dict(
        color='rgba(58, 71, 80, 0.6)',
        line=dict(color='rgba(58, 71, 80, 1.0)', width=3)
    )
))

fig2.add_trace(go.Bar(
    y=y2,
    x=[approvedLoan['response.loan.amount'].shape[0]],
    name='Accepted',
    base=0,
    orientation='h',
    marker=dict(
        color='rgba(58, 71, 80, 0.6)',
        line=dict(color='rgba(58, 71, 80, 1.0)', width=3)
    )
))

fig1.update_layout(
    barmode='stack', 
    title={'text': f"Accepted vs Rejected",
          'x':0.5,
          'xanchor': 'center'
    })

fig2.update_layout(
    barmode='stack', 
    title={'text': f"Accepted vs Rejected",
          'x':0.5,
          'xanchor': 'center'
    })

fig1.update_yaxes(
        ticktext=y1,
        tickvals=y1
    )

fig2.update_yaxes(
        ticktext=y2,
        tickvals=y2
    )

fig1.show()
fig2.show()
```
